```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

Q1: Berapa jumlah baris dan kolom dataset titanic? kolom apa saja yang tersedia?

```python
print(df.shape)
print(df.columns)
```

Data set titanic memiliki 891 baris dan 12 kolom yang merepresentasikan informasi penumpang seperti kelas, jenis kelamin, usia, dan status keselamatan.

Q2: kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)

```python
(df.isnull().sum() / len(df)) * 100
```

Terdapat missing value pada beberap kolom, terutama Cabin (687 data kosong), Age 177, dan Embarked. Kolom Cabin memiliki jumlah data kosong peling bnyak sehingga kemungkinan perlu di hapus atau tidak di gunakan dalam analisis

Q3: Berapa persen penumpang yang selamat (Survived=1)? gunakan oprasi Numphy

```python
round(df["Survived"].mean() * 100, 2)
```

presentase penumpang yang selamat adalah sekitar 38% dari total peenumpang

Q4: Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
wanita_kelas1 = df[(df["Sex"] == "female") & (df["Pclass"] == 1)]

print("Jumlah", len(wanita_kelas1))
print("Rata-rata umur:", wanita_kelas1["Age"].mean())
```

terdapat 94 penumpang wanita pada kelas 1 dengan rata-rata usia sekitar 34.6 tahun

Q5: Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas
penumpang. Kelas mana yang paling tinggi?

```python
df.groupby("Pclass")["Survived"].mean()
```

Tingkat keselamatan penumpang berbeda berdasarkan kelas. Penumpang kelas 1 memiliki tingkat keselamatan tertinggi (63%), diikuti kelas 2 (47%), dan kelas 3 (24%). Hal ini menunjukan bahwa kelas sosial mempengaruhi peluang keselamatan.

Q6: Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik.
Sertakan interpretasi singkat.

```python
df.groupby("Survived")["Fare"].mean()
```

Berdasarkan analisis harga tiket, penumpang yang selamat memiliki rata-rata harga tiket sebesar 48.39, lebih tinggi dibandingkan penumpang yang tidak selamat sebesar 22.11. Hal ini menunjukkan bahwa faktor ekonomi berpengaruh terhadap peluang keselamatan penumpang.
